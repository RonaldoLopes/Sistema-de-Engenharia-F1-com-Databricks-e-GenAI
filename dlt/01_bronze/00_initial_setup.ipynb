# # Configuração Inicial do Projeto F1 Databricks
# Este notebook configura o ambiente, catálogos e schemas necessários

```python
%sql
CREATE CATALOG IF NOT EXISTS f1_lakehouse
COMMENT 'Catálogo criado para armazenar bases de dados do projeto';
USE CATALOG f1_lakehouse;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze COMMENT 'Schema para dados bronze';
CREATE SCHEMA IF NOT EXISTS silver COMMENT 'Schema para dados silvera';
CREATE SCHEMA IF NOT EXISTS gold COMMENT 'Schema para dados gold';
```

```python
%sql
CREATE TABLE IF NOT EXISTS bronze.ingestion_control (
    source_name STRING,
    season INT,
    endpoint STRING,
    records_ingested INT,
    ingested_at TIMESTAMP,
    status STRING,
    error_message STRING
  ) USING DELTA
  TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true'
  );

CREATE TABLE IF NOT EXISTS bronze.drivers_raw (
    driverId STRING,
    driverRef STRING,
    number INT,
    code STRING,
    forename STRING,
    surname STRING,
    nationality STRING,
    dateOfBirth STRING,
    url STRING,
    _ingested_at TIMESTAMP,
    _source STRING,
    _season INT
  ) USING DELTA
  TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true'
  );

CREATE TABLE IF NOT EXISTS bronze.constructors_raw (
    constructorId STRING,
    constructorRef STRING,
    name STRING,
    nationality STRING,
    url STRING,
    _ingested_at TIMESTAMP,
    _source STRING,
    _season INT
  ) USING DELTA
  TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true'
  );

CREATE TABLE IF NOT EXISTS bronze.races_raw (
    raceId STRING,
    year INT,
    round INT,
    circuitId STRING,
    name STRING,
    date STRING,
    time STRING,
    url STRING,
    _ingested_at TIMESTAMP,
    _source STRING,
    _season INT
  ) USING DELTA
  TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true'
  );

CREATE TABLE IF NOT EXISTS bronze.results_raw (
    resultId STRING,
    raceId STRING,
    driverId STRING,
    constructorId STRING,
    number INT,
    grid INT,
    position INT,
    positionText STRING,
    positionOrder INT,
    points FLOAT,
    laps INT,
    time STRING,
    milliseconds INT,
    fastestLap INT,
    rank INT,
    fastestLapTime STRING,
    fastestLapSpeed FLOAT,
    statusId STRING,
    _ingested_at TIMESTAMP,
    _source STRING,
    _season INT
  ) USING DELTA
  TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true'
  );
```
